```python

#multilevel inheritance
class x():
    def fun(self):
        print("hii")
class y(x):
    def fun1(self):
        print("hello")
        
class z(y):
    def fun2(self):
        print("namaskara")   
        
object=z()
object.fun()
object.fun1()
object.fun2()
print()
                      
        
        
```

```python

#multiple inheritance

class x():
    def fun(self):
        print("hii")
class y():
    def fun1(self):
        print("hello")
        
class z(x,y):
    def fun2(self):
        print("namaskara")   
        
object=z()
object.fun()
object.fun1()
object.fun2()
print()

      
```

```python
#hybrid inheritance
class a():
    def fun(self):
        print("papas money")
        
class b():
    def fun1(self):
        print("my money")
                
class x(a,b):
    def fun2(self):
        pass
        
class y(a):
    def fun3(self):    #method
        pass   
        
x1=x()  #object
x1.fun()
x1.fun1()
y1=y()
y1.fun()

```
